```python
def intersect_area(top_left1, bottom_right1, top_left2, bottom_right2):
    # Check for intersection
    if top_left1[0] > bottom_right2[0] or top_left2[0] > bottom_right1[0] or top_left1[1] > bottom_right2[1] or top_left2[1] > bottom_right1[1]:
        return 0  # No intersection, area is 0

    # Calculate intersection coordinates
    x1 = max(top_left1[0], top_left2[0])
    y1 = max(top_left1[1], top_left2[1])
    x2 = min(bottom_right1[0], bottom_right2[0])
    y2 = min(bottom_right1[1], bottom_right2[1])

    # Calculate intersection area
    intersection_area = max(0, (x2 - x1) * (y2 - y1))
    return intersection_area


intersect_area(top_left, bottom_right, top_left1, bottom_right2)
```

```python
def intersect_coordinates(top_left1, bottom_right1, top_left2, bottom_right2):
    # Check for intersection
    if top_left1[0] > bottom_right2[0] or top_left2[0] > bottom_right1[0] or top_left1[1] > bottom_right2[1] or top_left2[1] > bottom_right1[1]:
        return None  # No intersection

    # Calculate intersection coordinates
    x1 = max(top_left1[0], top_left2[0])
    y1 = max(top_left1[1], top_left2[1])
    x2 = min(bottom_right1[0], bottom_right2[0])
    y2 = min(bottom_right1[1], bottom_right2[1])

    # Calculate intersection area
    intersection_area = max(0, (x2 - x1) * (y2 - y1))

    # Return intersection coordinates and area
    return (x1, y1), (x2, y2), intersection_area

k = intersect_coordinates(top_left, bottom_right, top_left1, bottom_right2)
im2 = cv2.rectangle(img, k[0], k[1], (0, 0, 255), 3, )
# cv2_imshow(im2)
plt.imshow(im)
plt.axis('off')  # Hide axes
plt.show()
```
